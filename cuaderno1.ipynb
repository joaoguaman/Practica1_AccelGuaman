```python
from faker import Faker
fake = Faker()

lista = []

for i in range(0,10000):
    lista.append({'nombre':fake.name(),'apellido':fake.last_name(),'email':fake.email(),'direccion':fake.address()})    
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(lista)
```

```python
data
```

```python
import random

data['edad'] = [random.randint(30, 40) for _ in range(len(data))]
```

```python
data
```

```python
provincias = ['Loja', 'Pichincha', 'Guayas', 'Azuay', 'Manabi']

data['provincia'] = [random.choice(provincias) for _ in range(len(data))]
```

```python
data
```

```python
data.to_csv("data_cuaderno1.csv",index=False)
```
